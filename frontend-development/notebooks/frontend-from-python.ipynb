## 1️⃣ The DOM as a Tree (Python Dictionaries)

The **Document Object Model (DOM)** is just a tree of nested dictionaries in concept.

**JavaScript:**
```javascript
const element = {
  tagName: 'div',
  attributes: { id: 'container', class: 'main' },
  children: [...]
}
```

**Python equivalent:**

````python
# DOM as nested Python dictionaries
dom_tree = {
    'tag': 'html',
    'attrs': {},
    'children': [
        {
            'tag': 'head',
            'attrs': {},
            'children': [
                {
                    'tag': 'title',
                    'attrs': {},
                    'children': ['My Page']
                }
            ]
        },
        {
            'tag': 'body',
            'attrs': {'class': 'container'},
            'children': [
                {
                    'tag': 'h1',
                    'attrs': {'id': 'title'},
                    'children': ['Welcome!']
                },
                {
                    'tag': 'p',
                    'attrs': {},
                    'children': ['This is a paragraph.']
                },
                {
                    'tag': 'ul',
                    'attrs': {},
                    'children': [
                        {'tag': 'li', 'attrs': {}, 'children': ['Item 1']},
                        {'tag': 'li', 'attrs': {}, 'children': ['Item 2']},
                        {'tag': 'li', 'attrs': {}, 'children': ['Item 3']},
                    ]
                }
            ]
        }
    ]
}

print("DOM Tree structure:")
import json
print(json.dumps(dom_tree, indent=2))
````

### DOM Traversal (Like Walking a Tree)

Just like traversing nested Python data structures:

````python
def find_element_by_id(node, target_id):
    """Like document.getElementById() in JavaScript"""
    if isinstance(node, dict):
        if node.get('attrs', {}).get('id') == target_id:
            return node
        
        for child in node.get('children', []):
            result = find_element_by_id(child, target_id)
            if result:
                return result
    return None

def find_elements_by_tag(node, tag_name):
    """Like document.getElementsByTagName() in JavaScript"""
    results = []
    
    if isinstance(node, dict):
        if node.get('tag') == tag_name:
            results.append(node)
        
        for child in node.get('children', []):
            results.extend(find_elements_by_tag(child, tag_name))
    
    return results

# Test it out
title_element = find_element_by_id(dom_tree, 'title')
print(f"Found element with id='title': {title_element}\n")

all_li_elements = find_elements_by_tag(dom_tree, 'li')
print(f"Found {len(all_li_elements)} <li> elements:")
for li in all_li_elements:
    print(f"  - {li['children'][0]}")
````

## 2️⃣ Event Loop: asyncio vs JavaScript

JavaScript's event loop is conceptually similar to Python's `asyncio`.

**Key difference:** JavaScript is single-threaded with async I/O by default. Python uses explicit `async`/`await`.

### JavaScript Event Loop (Conceptual):
```javascript
console.log('1');
setTimeout(() => console.log('2'), 0);
console.log('3');
// Output: 1, 3, 2
```

### Python asyncio equivalent:

````python
import asyncio
from datetime import datetime

async def simulate_event_loop():
    """Simulate JavaScript's event loop behavior"""
    
    print(f"[{datetime.now().strftime('%H:%M:%S.%f')[:-3]}] 1. Synchronous code")
    
    # setTimeout equivalent - schedule for later
    asyncio.create_task(delayed_task(0, "2. Async callback (like setTimeout)"))
    
    print(f"[{datetime.now().strftime('%H:%M:%S.%f')[:-3]}] 3. More synchronous code")
    
    # Give event loop time to process
    await asyncio.sleep(0.1)

async def delayed_task(delay, message):
    await asyncio.sleep(delay)
    print(f"[{datetime.now().strftime('%H:%M:%S.%f')[:-3]}] {message}")

# Run the event loop
await simulate_event_loop()

print("\n🔍 Notice: Just like JavaScript, async code runs after sync code!")
````

### Multiple Concurrent Tasks (Like Multiple API Calls)

````python
async def fetch_data(api_name, delay):
    """Simulate API call (like fetch() in JavaScript)"""
    print(f"📡 Requesting {api_name}...")
    await asyncio.sleep(delay)
    print(f"✅ {api_name} completed after {delay}s")
    return {'api': api_name, 'data': f'Data from {api_name}'}

async def fetch_all_parallel():
    """Like Promise.all() in JavaScript"""
    print("Starting parallel requests...\n")
    
    # All start at the same time (non-blocking)
    results = await asyncio.gather(
        fetch_data('User API', 0.3),
        fetch_data('Products API', 0.5),
        fetch_data('Analytics API', 0.2),
    )
    
    print(f"\n📦 All results: {results}")

await fetch_all_parallel()

print("\n💡 In JavaScript: Promise.all([fetch(...), fetch(...), fetch(...)])")
````

## 3️⃣ JSON & HTTP: requests vs fetch()

**Python (requests):**
```python
import requests
response = requests.get('https://api.example.com/data')
data = response.json()
```

**JavaScript (fetch):**
```javascript
const response = await fetch('https://api.example.com/data');
const data = await response.json();
```

Let's simulate both patterns:

````python
import json
from typing import Dict, Any

# Mock API response
MOCK_API_DATA = {
    'users': [
        {'id': 1, 'name': 'Alice', 'email': 'alice@example.com'},
        {'id': 2, 'name': 'Bob', 'email': 'bob@example.com'},
    ],
    'status': 'success'
}

# Python style (synchronous)
def fetch_sync(url: str) -> Dict[str, Any]:
    """Python requests-style API call (blocking)"""
    print(f"🐍 Python fetch (blocking): {url}")
    # Simulate network delay
    import time
    time.sleep(0.1)
    return MOCK_API_DATA

# JavaScript style (async)
async def fetch_async(url: str) -> Dict[str, Any]:
    """JavaScript fetch-style API call (non-blocking)"""
    print(f"⚛️ JavaScript fetch (non-blocking): {url}")
    await asyncio.sleep(0.1)  # Simulate network delay
    return MOCK_API_DATA

# Comparison
print("=== Python Style ===")
data = fetch_sync('/api/users')
print(f"Got data: {data['users'][0]['name']}\n")

print("=== JavaScript Style ===")
data = await fetch_async('/api/users')
print(f"Got data: {data['users'][0]['name']}\n")

print("💡 Key difference: JavaScript fetch is async by default!")
````

## 4️⃣ Component Trees: Python Classes as Analogy

React components are like **Python classes** that render themselves.

**React Component:**
```javascript
function Button({ label, onClick }) {
  return <button onClick={onClick}>{label}</button>
}
```

**Python Class Analogy:**

````python
from dataclasses import dataclass
from typing import List, Callable, Optional

@dataclass
class Component:
    """Base class for all components"""
    
    def render(self) -> dict:
        """Return DOM structure (like JSX)"""
        raise NotImplementedError

@dataclass
class Button(Component):
    """Like a React Button component"""
    label: str
    on_click: Optional[Callable] = None
    
    def render(self) -> dict:
        return {
            'tag': 'button',
            'attrs': {'class': 'btn'},
            'children': [self.label],
            'events': {'click': self.on_click}
        }

@dataclass
class Card(Component):
    """Composite component (contains other components)"""
    title: str
    content: str
    actions: List[Component]
    
    def render(self) -> dict:
        return {
            'tag': 'div',
            'attrs': {'class': 'card'},
            'children': [
                {'tag': 'h2', 'attrs': {}, 'children': [self.title]},
                {'tag': 'p', 'attrs': {}, 'children': [self.content]},
                {
                    'tag': 'div',
                    'attrs': {'class': 'actions'},
                    'children': [action.render() for action in self.actions]
                }
            ]
        }

# Create component tree
def handle_save():
    print("💾 Save clicked!")

def handle_cancel():
    print("❌ Cancel clicked!")

card = Card(
    title="User Profile",
    content="Edit your profile information below.",
    actions=[
        Button(label="Save", on_click=handle_save),
        Button(label="Cancel", on_click=handle_cancel),
    ]
)

# Render the tree
rendered = card.render()
print("Component Tree:")
print(json.dumps(rendered, indent=2, default=str))

# Simulate clicking buttons
print("\nSimulating button clicks:")
for action in card.actions:
    if action.on_click:
        action.on_click()
````

## 5️⃣ State Management: dataclass vs useState

**React useState:**
```javascript
const [count, setCount] = useState(0);
```

**Python equivalent with state tracking:**

````python
from dataclasses import dataclass, field
from typing import Callable, List

@dataclass
class State:
    """Simple state container with change tracking (like React state)"""
    _value: any
    _listeners: List[Callable] = field(default_factory=list)
    _render_count: int = 0
    
    @property
    def value(self):
        return self._value
    
    def set_value(self, new_value):
        """Like setState in React - triggers re-render"""
        if self._value != new_value:
            old_value = self._value
            self._value = new_value
            self._render_count += 1
            
            print(f"🔄 State changed: {old_value} → {new_value}")
            print(f"📊 Render count: {self._render_count}")
            
            # Notify listeners (like React re-rendering)
            for listener in self._listeners:
                listener(new_value)
    
    def subscribe(self, listener: Callable):
        """Add a listener (like useEffect in React)"""
        self._listeners.append(listener)

@dataclass
class Counter:
    """Stateful component (like a React functional component with useState)"""
    
    def __post_init__(self):
        self.count = State(0)
        
        # Subscribe to state changes (like useEffect)
        self.count.subscribe(self._on_count_change)
    
    def _on_count_change(self, new_count):
        """Effect runs when count changes"""
        print(f"  ↳ Effect triggered: Count is now {new_count}")
    
    def increment(self):
        """Like: setCount(count + 1)"""
        self.count.set_value(self.count.value + 1)
    
    def decrement(self):
        """Like: setCount(count - 1)"""
        self.count.set_value(self.count.value - 1)
    
    def render(self) -> dict:
        """Render the component (returns DOM structure)"""
        return {
            'tag': 'div',
            'attrs': {'class': 'counter'},
            'children': [
                {'tag': 'h2', 'attrs': {}, 'children': [f'Count: {self.count.value}']},
                {'tag': 'button', 'attrs': {}, 'children': ['+'], 'events': {'click': self.increment}},
                {'tag': 'button', 'attrs': {}, 'children': ['-'], 'events': {'click': self.decrement}},
            ]
        }

# Test the counter
counter = Counter()

print("Initial state:")
print(f"Count: {counter.count.value}\n")

print("Simulating user interactions:\n")
counter.increment()  # Like clicking +
print()
counter.increment()  # Like clicking + again
print()
counter.decrement()  # Like clicking -
print()

print(f"\n📈 Final count: {counter.count.value}")
print(f"🔢 Total re-renders: {counter.count._render_count}")
````

## 6️⃣ Props vs Function Arguments

**React props** are just function arguments:

```javascript
function Greeting({ name, age }) {
  return <h1>Hello {name}, you are {age}</h1>
}

<Greeting name="Alice" age={30} />
```

**Python equivalent:**

````python
@dataclass
class Greeting(Component):
    """Component with props (just instance attributes)"""
    name: str
    age: int
    
    def render(self) -> dict:
        return {
            'tag': 'h1',
            'attrs': {},
            'children': [f'Hello {self.name}, you are {self.age}']
        }

@dataclass
class UserList(Component):
    """Component that maps over data (like .map() in React)"""
    users: List[dict]
    
    def render(self) -> dict:
        return {
            'tag': 'div',
            'attrs': {'class': 'user-list'},
            'children': [
                Greeting(name=user['name'], age=user['age']).render()
                for user in self.users
            ]
        }

# Create component with props
users_data = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35},
]

user_list = UserList(users=users_data)
rendered = user_list.render()

print("Rendered user list:")
print(json.dumps(rendered, indent=2))

print("\n💡 In React: users.map(user => <Greeting name={user.name} age={user.age} />)")
````

## 7️⃣ HTML Page Generator in Python

Let's build a complete HTML page generator (like JSX → HTML):

````python
def render_to_html(node) -> str:
    """Convert our component tree to actual HTML (like ReactDOM.render)"""
    
    # Handle text nodes
    if isinstance(node, str):
        return node
    
    # Handle component nodes
    if not isinstance(node, dict):
        return ''
    
    tag = node.get('tag', 'div')
    attrs = node.get('attrs', {})
    children = node.get('children', [])
    
    # Build attributes string
    attrs_str = ''.join([
        f' {key}="{value}"' for key, value in attrs.items()
    ])
    
    # Self-closing tags
    if tag in ['img', 'br', 'hr', 'input']:
        return f'<{tag}{attrs_str} />'
    
    # Render children
    children_html = ''.join([
        render_to_html(child) for child in children
    ])
    
    return f'<{tag}{attrs_str}>{children_html}</{tag}>'

# Create a full page
@dataclass
class App(Component):
    """Root application component"""
    
    def render(self) -> dict:
        return {
            'tag': 'html',
            'attrs': {},
            'children': [
                {
                    'tag': 'head',
                    'attrs': {},
                    'children': [
                        {'tag': 'title', 'attrs': {}, 'children': ['My App']},
                        {'tag': 'style', 'attrs': {}, 'children': [
                            'body { font-family: sans-serif; padding: 2rem; }'
                        ]}
                    ]
                },
                {
                    'tag': 'body',
                    'attrs': {},
                    'children': [
                        Card(
                            title="Python → React",
                            content="Now you understand React from a Python perspective!",
                            actions=[
                                Button(label="Learn More"),
                                Button(label="Build Something"),
                            ]
                        ).render()
                    ]
                }
            ]
        }

# Generate HTML
app = App()
html = render_to_html(app.render())

print("Generated HTML:")
print(html)

# Save to file
with open('/tmp/generated-page.html', 'w') as f:
    f.write(html)

print("\n✅ Saved to /tmp/generated-page.html")
print("💡 This is essentially what React does: Components → HTML")
````

## 🎓 Key Takeaways

| Concept | Python Equivalent | React/JavaScript |
|---------|-------------------|------------------|
| **DOM** | Nested dictionaries | Tree of objects |
| **Event Loop** | `asyncio` | Built-in (single-threaded) |
| **Components** | Classes with `render()` | Functions returning JSX |
| **Props** | Function arguments / `__init__` params | Component props |
| **State** | Instance variables | `useState` hook |
| **Effects** | Property setters / decorators | `useEffect` hook |
| **API Calls** | `requests` (blocking) | `fetch` (async by default) |
| **Rendering** | String generation | Virtual DOM diffing |

### What's Different?

1. **Reactivity:** React automatically re-renders on state change. Python requires manual updates.
2. **Async by Default:** JavaScript I/O is async. Python I/O is sync unless you use `asyncio`.
3. **Virtual DOM:** React optimizes updates. Python templates regenerate everything.
4. **Type System:** TypeScript is optional but recommended. Python has type hints.
